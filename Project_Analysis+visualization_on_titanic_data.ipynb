```python
#From where max & min passenger board to titanic and also define their avg age

df2[['Embarked','Fare']].aggregate([min,max]),df2['Age'].mean()

# Coclusion----->> In this there are minimum passenger board from the 'C' Embarked and maximum passenger board from the 'S' and the avg of 'AGE' is 29.58
```

```python
# tell me the name of the youngest passenger on titanic.

youngest_and_oldest_Passenger=df2[['Name','Age']].aggregate([min,max])
youngest_and_oldest_Passenger


#Conclusion---->> in this data the youngest passenger age is "0.42" and oldest passenger age is "80.00" and shown in bellow graph
```

```python
sns.barplot('Name','Age',data=youngest_and_oldest_Passenger)
```

```python
# what is the highest & lowest fare on titanic.


Min_and_max_Fare_of_passenger=df2[['Cabin','Fare']].aggregate([min,max])
Min_and_max_Fare_of_passenger


# Conclussion---->> in this data the minimum Fare is '0.00' of 'A10' Cabin and the maximum Fare is '512.32' of 'T Cabin'
```
